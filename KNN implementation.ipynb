```python
import matplotlib.pyplot as plt
import numpy as np
from math import sqrt
from matplotlib import style
from collections import Counter
import pandas as pd
```

```python
dataset = {'k':[[1,2],[2,3],[3,1],[1,4],[4,5]], 'r':[[7,4],[8,6],[8,9],[7,8],[6,9],[6,6],[5,8]]}
new_features = [5,7]
```

```python
style.use('ggplot')
```

```python
for i in dataset:
    for j in dataset[i]:
        plt.scatter(j[0], j[1], s=100, color=i)

plt.title('Our Data')
plt.xlabel('Axis X')
plt.ylabel('Axis Y')
plt.scatter(new_features[0], new_features[1],color='b')
plt.annotate('Our new dot', xy=(new_features[0], new_features[1]), xytext=(4, 1.5),arrowprops=dict(facecolor='blue', shrink=0.05),)

plt.draw()
```

```python
def k_nearestneighbors(data, predict, k):
    distance = []
    for group in data:
        for features in data[group]:
            #euclidean_distance = np.sqrt((features[0]-predict[0]**2) + (features[1]-prdict[1]**2))
            euclidean_distance = np.linalg.norm(np.array(features)-np.array(predict))
            distance.append([euclidean_distance, group])
    votes = [i[1] for i in sorted(distance)[:k]]
    
    print(Counter(votes).most_common(1))
    vote_result = Counter(votes).most_common(1)[0][0]
    
    return vote_result
```

```python
result = k_nearestneighbors(dataset, new_features, k=3)
print(result)
```

```python
with plt.xkcd():
    for i in dataset:
        for ii in dataset[i]:
            plt.scatter(ii[0], ii[1], s=100, color=i)
            
    plt.title('Our Data')
    plt.xlabel('Axis X')
    plt.ylabel('Axis Y')
    plt.scatter(new_features[0], new_features[1],color=result)
    plt.annotate('Our new dot', xy=(new_features[0], new_features[1]), xytext=(4, 1.5),arrowprops=dict(facecolor='blue', shrink=0.05),)

    plt.show()
```
